```python
import pickle
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import random
import pandas as pd

%matplotlib inline
```

```python
DATADIR = '/home/rohit/face_recog_coursera/students_images'
CATEGORIES = []
for people in os.listdir(DATADIR):
    CATEGORIES.append(people)
CATEGORIES
```

```python
full_dataset = []
img_size = 96
for category in CATEGORIES:
    path = os.path.join(DATADIR,category)
    class_num = CATEGORIES.index(category)
    for img in os.listdir(path):
        img_converted = cv2.imread(os.path.join(path,img))
        img_converted = cv2.resize(img_converted, (img_size,img_size))
        img_converted = cv2.cvtColor(img_converted, cv2.COLOR_BGR2RGB)
        full_dataset.append([img_converted,class_num])
```

```python
k = random.randint(0, len(full_dataset))
print('This is image of',CATEGORIES[full_dataset[k][1]])
plt.imshow(full_dataset[k][0])
plt.show()
```

```python
random.shuffle(full_dataset)
```

```python
X = []
Y = []

for img,labels in full_dataset:
    X.append(img)
    Y.append(labels)
X = np.array(X)
Y = np.array(Y)
print(Y)
```

```python
k = random.randint(0, len(full_dataset))
print('This is image of',CATEGORIES[Y[k]])
plt.imshow(X[k])
plt.show()
```

```python
pickle_out = open('X.pickle','wb')
pickle.dump(X,pickle_out)
pickle_out.close()

pickle_out = open('Y.pickle','wb')
pickle.dump(Y,pickle_out)
pickle_out.close()
```
